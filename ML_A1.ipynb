```python
import numpy as np
import pandas as pd

#importing the dataset from .csv file
df = pd.read_csv('iris.csv', names =['x1','x2','x3','x4','y'], skiprows = 1)
y1 = df['y'].values
# converting the labels to integer values.
label = {'Iris-setosa': 1,'Iris-versicolor': 2,'Iris-virginica': 3}
y1 = [label[item] for item in y1]

x11 = df['x1'].values
x22 = df['x2'].values 
x33 = df['x3'].values 
x44 = df['x4'].values 

#initialize X with parameters:x11,x22,x33,x44
X = np.matrix([np.ones(150),x11,x22,x33,x44]).T

#initialize Y with labels
Y = np.matrix(y1).T


#Finding the b values using b=(X’X)^-1(X’Y)
B=np.dot((np.dot(X.T,X)).I,(np.dot(X.T,Y)))
print("****************************************************************")
print("Beta Values{bo,b1,b2,b3,b4}:")
print(B)
print("****************************************************************")

#initialize Y_caps = Xb
Y_caps = np.dot(X,B)


XTrain = []
YTrain = []
XTest = []
YTest = []
accuracy = []

#Splitting the data for cross-validation
SplitX = np.split(X,5)
SplitY = np.split(Y,5)

# K-fold for k=5
for i in range(0,len(SplitX)):
    for j in range(0,len(SplitY)):
        if j == i:
            XTest.append(SplitX[j])
            YTest.append(SplitY[j])
        else:
            XTrain.append(SplitX[j])
            YTrain.append(SplitY[j])
    # data splitting for training and testing
    XT=np.matrix(np.concatenate(np.array(XTrain)))
    YT=np.matrix(np.concatenate(np.array(YTrain)))
    XTT=np.matrix(np.array(XTest))
    YTT=np.matrix(np.array(YTest)).flatten()

    #finding the b,y_cap and error for each fold.
    b=np.dot((np.dot(XT.T,XT)).I,(np.dot(XT.T,YT)))
    y_caps = np.dot(XTT,b).T.round()
    error=(YTT - y_caps)
    e=error.astype(int)
    #finding Root-mean-square Error for the fold
    rmse =np.sqrt(np.sum(np.square(YTT - y_caps))/30)  
    print("****************************************************************")
    print("fold:",i+1)
    print("Shape of Y(actual):",YTT.shape)
    print(YTT)
    print("Shape of X:",XTT.shape)
    print("Shape of b:",b.shape)
    print("Shape of Y_cap(Predicted):",y_caps.shape)
    print(y_caps.astype(int))
    
    accuracy.append(1-(np.count_nonzero(e)/30))
    print("Accuracy after each fold")
    print(accuracy)
    print("Root-mean-square Error for the fold:",rmse)
    
    error=np.array([])
    XTrain.clear()
    YTrain.clear()
    XTest.clear()
    YTest.clear()
    
    #finding accuracy and error for the system
    avg = sum(accuracy)/len(accuracy)*100
    eavg = 100 - avg
    print("****************************************************************")
print("System's accuracy(%):",avg)
print("System's error(%):",eavg)
```
